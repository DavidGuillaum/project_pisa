## Dataset Overview

The [PISA](https://www.oecd.org/pisa/) project is led by the OECD (Organization for Economic Co-operation and Development). PISA stands for Program for International Student Assessment and the goal is to retrieve data about the education level. The gathering of data (with a survey of 510 000 students (Sample)) is done every 3 years. The data frame weâ€™ll deal with is from 2012.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")

base_color = sb.color_palette()[0]
```

```python
df_dig = pd.read_csv('df_dig.csv', encoding ='latin-1')
df_lang = pd.read_csv('df_lang.csv', encoding ='latin-1')
df_pBack = pd.read_csv('df_pBack.csv', encoding ='latin-1')
```

# Digitalization interest

>  Relationship between **math score** and **time of internet use** (on a log scale)  
> Key insight: **Negative** relationship between **math score** and **time of use**
 

```python
plt.figure(figsize = [10, 10])
sb.regplot(data = df_dig, x = 'timeOFUse', y = 'math')
plt.title("Relation between time of IT use and Math score")
plt.xlabel('time of internet use')
plt.ylabel('math score');
```

# Language interest

> Relationship between **being bilingual** and **read score**   
> Key insight: No significative relationship found between **being bilingual** and **read score**

```python
plt.figure(figsize = [10, 10])
sb.boxplot(data=df_lang, y='read', x='bilingual', color=base_color)
plt.title("Relation between being bilingual and read score")
plt.xlabel('being bilingual or more')
plt.ylabel('read score');
```

# Parent background interest

> Relationship between **parent education level** and **science score**   
> Key insight: **Positive** relationship found between **parent education level** and **science score** 

```python
plt.figure(figsize = [10, 10])

plt.subplot(1, 2, 1)
plt.title("Relation between parent education level and science score")
sb.regplot(data = df_pBack, x = 'parentLevel', y = 'science', x_jitter=0.04, scatter_kws={'alpha':1/10}, fit_reg=False)
plt.xlabel('Parent education level')
plt.ylabel('Science score');
plt.subplot(1, 2, 2)
plt.hist2d(data = df_pBack, x = 'parentLevel', y = 'science')
plt.colorbar()
plt.xlabel('Parent education level')
plt.ylabel('Science score');
```
